# Pré-processamento

## Remoção de "Stop Words"

Redução de dimensionalidade com palavras inuteis

### Manual

```python
stopwords = ['a', 'agora', 'algum', 'alguma', 'aquele', 'aqueles', 'de', 'deu', 'do', 'e', 'estou', 'esta', 'esta',
             'ir', 'meu', 'muito', 'mesmo', 'no', 'nossa', 'o', 'outro', 'para', 'que', 'sem', 'talvez', 'tem', 'tendo',
             'tenha', 'teve', 'tive', 'todo', 'um', 'uma', 'umas', 'uns', 'vou']
```

```python
def removestopwords(texto):
    
    frases = []
    
    for (palavras,emocao) in texto:
        
        semstop = [p for p in palavras.split() if p not in stopwords] # Removendo stopwords.
        
        frases.append((semstop,emocao))
        
    return frases
```

```python
removestopwords(base)
```

## Steeming

Extração de radicais das palavras (raiz da palavra)

Redução da dimensionalidade da base de dados

Pode **podar** as infos até demais 

```python
def aplicastemmer(texto):
    
    stemmer = nltk.stem.RSLPStemmer() # compatível com portugues
    
    frases_stemming = []
    
    for (palavras,emocao) in texto:
        
        comstemming = [str(stemmer.stem(p)) for p in palavras.split() if p not in stopwordsnltk] # Aplica stemmer e também retira stopwords
        
        frases_stemming.append((comstemming,emocao))
        
    return frases_stemming
```

```python
frases_c_steeming = aplicastemmer(base)
```

## Listagem de palavras na Base

```python
def buscapalavras(frase):
    todaspalavras=[]
    
    for (palavras,emocao) in frase:
        todaspalavras.extend(palavras) # extend mescla as palavras cria uma lista com as listas passadas no extend
        
    return todaspalavras
```

## Extração de Palavras únicas 

### Extração de Frequência das Palavras

```python
def buscafrequencia(palavras): # parâmetro lista de palavras 
    
    palavras = nltk.FreqDist(palavras)
    
    return palavras
```

```python
# Lista de palavras únicas

def buscapalavrasunicas(frequencia):
    
    freq = frequencia.keys() # Já retorna valores únicos de "chaves" ou seja remove duplicados;
    
    return freq
```

```python
palavras_unicas = buscapalavrasunicas(frequencia)
palavras_unicas #Essa será já a lista final de palavras que irão binarizar pro treino.
```

## Extração de conteudo de palavras:
    
Dado uma frase, checo na lista de palavras únicas se ela existe e crio um flag apontando sobre isso

### Manual

```python
def extrator_palavras(documento):
    
    doc = set(documento)
    
    caract = {} # dicionário que vai dizer se existe ou não a palavra
    
    for palavra in palavras_unicas:
        
        caract['%s' % palavra] = (palavra in doc) # Palavra(chave) : Bool
        
    return caract
```

```python
caract_frase = extrator_palavras(['am','nov','dia'])
caract_frase
```

### Automático

```python
basecompleta = nltk.classify.apply_features(extrator_palavras,frases_c_steeming) # ela pega cada uma das frases e aplica no extrator
```
